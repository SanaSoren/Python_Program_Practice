# Write a program to create an Indian Kirana Store Calculator

```python
sum = 0
while (True):
    UserInput = input("Enter the product price or press the q to quit: \n")
    if (UserInput !='q'):
        sum = sum + int(UserInput)
        print(f"Order total so far: {sum}")
    else:
        print(f"Sum Total of the product {sum}. Thanks for shopping with us ")
        break
    
```
